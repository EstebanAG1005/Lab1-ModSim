# Ejercicio 5

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfinv

uniform_sample = np.random.uniform(size=500)

inverse_transform_sample = np.sqrt(2) * erfinv(2 * uniform_sample - 1)

plt.hist(inverse_transform_sample, bins=30, density=True)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histograma de una distribución normal generada con una muestra de transformación inversa')
plt.show()
```

```python
uniform_sample = np.random.uniform(size=500)

inverse_transform_sample = np.sqrt(2) * erfinv(2 * uniform_sample - 1)

plt.hist(inverse_transform_sample, bins=30, density=True, alpha=0.7, label='Histogram')

x = np.linspace(min(inverse_transform_sample), max(inverse_transform_sample), 500)
pdf_values = np.exp(-x**2/2) / np.sqrt(2*np.pi)

plt.plot(x, pdf_values, color='red', linewidth=2, label='PDF')

plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Histograma y PDF de una distribución normal generada con una muestra de transformación inversa')
plt.legend()
plt.show()
```

# Ejercicio 6

```python
from scipy.stats import gamma

forma = 2
escala = 3

muestra_gamma = np.random.gamma(forma, escala, size=1000)

shape_est, loc_est, scale_est = gamma.fit(muestra_gamma)

print("Parámetros estimados:")
print(f"Forma: {shape_est}")
print(f"Escala: {scale_est}")

print(f"Acercamiento forma {(forma - shape_est) / forma}")
print(f"Acercamiento escala {(escala - scale_est) / escala}")
```


# Ejercicio 7

```python
from scipy.stats import kstest

gamma_est = gamma(shape_est, loc_est, scale_est)

kstest_result = kstest(muestra_gamma, gamma_est.cdf)

print("Resultado de la prueba de Kolmogorov-Smirnov:")
print(f"Estadístico D: {kstest_result.statistic}")
print(f"P-valor: {kstest_result.pvalue}")
```

```python
mean = [0, 0]
cov = [[1, 0.5], [0.5, 1]]

data = np.random.multivariate_normal(mean, cov, size=1000)

x = data[:, 0]
y = data[:, 1]

plt.scatter(x, y, alpha=0.5)
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Diagrama de dispersión de la distribución normal bivariada')
plt.show()

correlation_coef = np.corrcoef(x, y)[0, 1]
print(f"Coeficiente de correlación: {correlation_coef}")
```

```python
from scipy.stats import chisquare

data = np.random.uniform(0, 1, 500)

bins = np.linspace(0, 1, 11)
hist, _ = np.histogram(data, bins=bins)

expected = np.full_like(hist, 500/(len(bins)-1))

chi2, p_value = chisquare(hist, expected)

print(f"bondad de ajuste de chi-cuadrado: {chi2}")
print(f"Valor-P: {p_value}")

# Interpretación
if p_value < 0.05:
    print("Rechazamos la hipótesis nula, los datos no siguen una distribución uniforme.")
else:
    print("No rechazamos la hipótesis nula, los datos podrían seguir una distribución uniforme.")

```
