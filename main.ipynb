# Ejercicio 1 - Propiedades de Distribución de Probabilidad

```python
#Genere una muestra aleatoria de 1000 puntos de datos de una distribución normal con media 0 y desviación estándar 1.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Muestra Aleatoria
data = np.random.normal(0, 1, 1000)

# Calcule e imprima la media, la varianza y la asimetría de la muestra.
mean = np.mean(data)
variance = np.var(data)
skewness = skew(data)

print(f"Media: {mean}")
print(f"Varianza: {variance}")
print(f"Asimetría: {skewness}")

# Trace un histograma de los datos y observe su forma.
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histograma de la distribución normal")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

#### ¿Qué le dicen estas estadísticas sobre la tendencia central, la dispersión y la forma de los datos?

Media: La media proporciona la tendencia central de los datos. ya que usamos una distribución normal con media 0, la media calculada debe estar cerca de 0.

Dispersión: Dado que generamos los datos con desviación estándar 1, la varianza debe estar cerca de 1.

Asímetría: Para una distribución normal, la asimetría es 0, por lo que el valor calculado debe estar cerca de 0.
