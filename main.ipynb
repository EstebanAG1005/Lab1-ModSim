### Ejercicio 1 - Propiedades de Distribución de Probabilidad

```python
#Genere una muestra aleatoria de 1000 puntos de datos de una distribución normal con media 0 y desviación estándar 1.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Muestra Aleatoria
data = np.random.normal(0, 1, 1000)

# Calcule e imprima la media, la varianza y la asimetría de la muestra.
mean = np.mean(data)
variance = np.var(data)
skewness = skew(data)

print(f"Media: {mean}")
print(f"Varianza: {variance}")
print(f"Asimetría: {skewness}")

# Trace un histograma de los datos y observe su forma.
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histograma de la distribución normal")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

### Ejercicio 2 - PDF & CDF


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Genere una muestra aleatoria de 1000 puntos de datos a partir de una distribución exponencial con parámetro de tasa 0.5.
scale = 1 / 0.5
data = expon.rvs(scale=scale, size=1000)

# Crea un rango de valores x para evaluar las funciones PDF y CDF
x = np.linspace(expon.ppf(0.01, scale=scale), expon.ppf(0.99, scale=scale), 1000)

# PDF
pdf = expon.pdf(x, scale=scale)

# CDF
cdf = expon.cdf(x, scale=scale)

# Trace la función de densidad de probabilidad (PDF) y la función de distribución acumulativa (CDF) de los datos.
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, pdf, 'r-', label='PDF')
plt.title('PDF')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, 'b-', label='CDF')
plt.title('CDF')
plt.grid(True)

plt.tight_layout()
plt.show()

# Calcule e imprima las probabilidades para intervalos específicos utilizando la CDF
# Intervalo [1, 2]
prob_interval_1 = expon.cdf(2, scale=scale) - expon.cdf(1, scale=scale)
print("La probabilidad para el intervalo [1, 2] es:", prob_interval_1)

# Intervalo [3, 4]
prob_interval_2 = expon.cdf(4, scale=scale) - expon.cdf(3, scale=scale)
print("La probabilidad para el intervalo [3, 4] es:", prob_interval_2)
```

### Ejercicio 3 - Bias & Kurtosis


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, skew, kurtosis

# Genere una muestra aleatoria de 1000 puntos de datos a partir de una distribución beta con parámetros de forma (2, 2).
shape_params = (2, 2)
sample = beta.rvs(*shape_params, size=1000)

# Calcule e imprima el sesgo y la curtosis de la muestra.
sample_skew = skew(sample)
sample_kurtosis = kurtosis(sample)

print(f'Sesgo (skew): {sample_skew}')
print(f'Curtosis (kurtosis): {sample_kurtosis}')

# Graficar la distribución de la muestra
plt.hist(sample, bins=30, density=True)
plt.title('Distribución de la Muestra')
plt.show()
```

### Ejercicio 4 - Generación de Números aleatorios

```python
import numpy as np
import matplotlib.pyplot as plt

# Genere una muestra aleatoria de 1000 puntos de datos a partir de una distribución de Poisson con parámetro lambda 5.
lambda_val = 5
sample = np.random.poisson(lambda_val, 1000)

#  Trace un histograma de los datos generados.
plt.hist(sample, bins=np.arange(sample.min(), sample.max()+1), alpha=0.6, color='g')

plt.title('Histograma de la muestra aleatoria de una distribución de Poisson')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.show()

# Calcule e imprima la media y la varianza de la muestra.
mean = np.mean(sample)
variance = np.var(sample)

print(f"Media de la muestra: {mean}")
print(f"Varianza de la muestra: {variance}")
```
